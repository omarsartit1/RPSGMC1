```python
import tkinter as tk 
import random
import sys 

# Prompt choices
def play():
    choices = ["rock", "paper", "scissors"]
    index = random.randint(0,2)
    computer_choice = choices[index]
    
    user_choice = user_entry.get().lower()

    if user_choice not in choices : 
        result.set("Invalid choice")
    elif user_choice == computer_choice :
        result.set("It's a tie !") 
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "scissors" and computer_choice == "paper") or \
         (user_choice == "paper" and computer_choice == "rock"):
        result.set("You win !")
    else :
        result.set("You lost !") 

def reset():
    reset_window = tk.Toplevel(window)
    reset_label = tk.Label(reset_window, text="Do you want to play again? (yes/no): ")
    reset_label.pack()
    reset_entry = tk.Entry(reset_window)
    reset_entry.pack()
    
    def reset_play():
        reset_choice = reset_entry.get().lower()
        if reset_choice == "yes":
            user_entry.delete(0, tk.END) 
            play()
            reset_window.destroy()
            result.set("")
        elif reset_choice == "no":
            Exit()
        else:
            result.set("Invalid input. Please enter 'yes' or 'no'.")
    reset_button = tk.Button(reset_window, text="Submit", command=reset_play)
    reset_button.pack()

def Exit():
    print("Thanks for playing! Goodbye!")
    window.destroy()

# GUI
window = tk.Tk()
window.title("Rock Paper Scissor")
window.configure(background='#0000ff')  # Blue background

# Create main frames
title_frame = tk.Frame(window, bg="blue")
title_frame.pack(pady=20)

entry_frame = tk.Frame(window, bg="blue")
entry_frame.pack(pady=20)

button_frame = tk.Frame(window, bg="blue")
button_frame.pack(pady=20)

result_frame = tk.Frame(window, bg="blue")
result_frame.pack(pady=20)

# Add widgets to frames
label = tk.Label(title_frame, text="ROCK PAPER SCISSOR", foreground="white", background="blue", font=("Helvetica", 24, "bold"))
label.pack()

user_entry = tk.Entry(entry_frame, width=20)
user_entry.pack() # entry user

play_button = tk.Button(button_frame, text="Play", command=play)
play_button.pack(side=tk.LEFT, padx=10)

reset_button = tk.Button(button_frame, text="Reset", command=reset)
reset_button.pack(side=tk.LEFT, padx=10)

result = tk.StringVar()
result_label = tk.Label(result_frame, textvariable=result, foreground="white", background="blue", font=("Helvetica", 18))
result_label.pack()

window.resizable(width=1000, height=1000)
window.mainloop()

 
```
